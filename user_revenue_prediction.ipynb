```python
import pandas as pd
from utils import UserRevenuePredictor, visualize_true_pred

SEED = 783403
DATA_LOCATION = "data"
MODELS_LOCATION = "models"
```

# Example dataframe

```python
# Get df
df = pd.read_csv(f"{DATA_LOCATION}/TEST TASK (dataset).zip")

# Drop target columns
X = df.drop(columns=["target_full_ltv_day30", "target_sub_ltv_day30",
                     "target_iap_ltv_day30", "target_ad_ltv_day30"])

# Get subdataframe for an example
df = df.sample(50, random_state=SEED)

# Get X and y
y = df["target_full_ltv_day30"].values
X = df.drop(columns=["target_full_ltv_day30", "target_sub_ltv_day30",
                     "target_iap_ltv_day30", "target_ad_ltv_day30"])
```

# Example of usage

Ми прописали готовий клас, який можна використовувати для предікту доходу на нових входженнях.

При ініціалізації класу можна передати два параметри:
- `bias_target` - булева змінна, яка вказує на те, чи використовувати модель із зміщеним таргетом, чи ні. За замовчуванням значення цього параметру `False`, тобто буде використовуватися наша основна модель із звичайним таргетом.
- `models_location` - шлях до папки із натренованими моделями. За замовчуванням `models`.

$\color{red}{\text{ВАЖЛИВО!}}$ Датафрейм, який передається у метод `predict` має бути у точності такий самий, який ми використовували під час побудови моделі, включно із к-стю колонок, їх назвами та їх порядком. Зрозуміло, що це не надто раціонально, проте дописати функцію для аналізу вхідного датафрейму можна в будь-який час, якщо краще розуміти, які дані і в якому вигляді будуть частіше передаватися на вхід.

```python
predictor = UserRevenuePredictor(bias_target=False, models_location=MODELS_LOCATION)
```

```python
y_pred = predictor.predict(X)
```

Якщо відомі справжні значення, можна побудувати графік залежності реальних даних доходів від спрогнозованих.

```python
visualize_true_pred(y, y_pred)
```
